```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("cumulative.csv")
```

```python
df = df.fillna(df.mean(numeric_only=True))
```

```python
df.drop(
    ["kepid", "kepoi_name"],
    axis=1,
    errors="ignore",
    inplace=True
)
```

```python
df["koi_disposition"] = df["koi_disposition"].map({
    "CONFIRMED": 1,
    "FALSE POSITIVE": 0,
    "CANDIDATE": 0
})
```

```python
df = pd.get_dummies(df, dtype=float)
```

```python
X = df.drop("koi_disposition", axis=1).astype(float).values
y = df["koi_disposition"].values
```

```python
X = (X - X.mean(axis=0)) / X.std(axis=0)
```

```python
split = int(0.8 * len(X))

X_train = X[:split]
X_test = X[split:]

y_train = y[:split]
y_test = y[split:]
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))


```

```python
weights = np.zeros(X_train.shape[1])
bias = 0

learning_rate = 0.01
epochs = 1000
```

```python
for _ in range(epochs):

    linear = np.dot(X_train, weights) + bias

    predictions = sigmoid(linear)

    dw = (1 / len(X_train)) * np.dot(
        X_train.T,
        (predictions - y_train)
    )

    db = (1 / len(X_train)) * np.sum(
        predictions - y_train
    )

    weights -= learning_rate * dw
    bias -= learning_rate * db
```

```python
linear = np.dot(X_test, weights) + bias

probabilities = sigmoid(linear)

predictions = (probabilities >= 0.5).astype(int)
```

```python
accuracy = np.mean(predictions == y_test)

print("Accuracy:", accuracy)
```
